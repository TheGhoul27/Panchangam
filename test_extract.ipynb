```python
from bs4 import BeautifulSoup as bs
import requests as req
from tabulate import tabulate
import datetime
import json

#css = "td+ td"
css = "td"
url = "https://www.panchangam.org/global/daily.php?city=New+York#google_vignette"

day_map = {
    'Monday': 'Indu Vasara',
    'Tuesday': 'Bhauma Vasara',
    'Wednesday': 'Saumya Vasara',
    'Thursday': 'Guru Vasara',
    'Friday': 'Bhrigu Vasara',
    'Saturday': 'Sthira Vasara',
    'Sunday': 'Bhanu Vasara'
}

ruthu_map = {
    'Chaithra': 'Vasanta Ruthu',
    'Vaisakha': 'Vasanta Ruthu',
    'Jyeshta': 'Greeshma Ruthu',
    'Ashadha': 'Greeshma Ruthu',
    'Sravana': 'Greeshma Ruthu',
    'Bhadrapada': 'Greeshma Ruthu',
    'Ashvina': 'Sharada Ruthu',
    'Kartika': 'Sharada Ruthu',
    'Agrahayana': 'Hemanta Ruthu',
    'Pausa': 'Hemanta Ruthu',
    'Magha': 'Shishira Ruthu',
    'Phalguna': 'Shishira Ruthu'
}

css2 = '.col-md-8 .panel-body .btn'
def get_cities(url):
    page = req.get(url)
    soup = bs(page.content, "html.parser")
    cities = soup.select(css2)
    return cities

def get_panchangam(url):
    page = req.get(url)
    soup = bs(page.content, "html.parser")
    panchangam = soup.select(css)
    return panchangam

panchangam = get_panchangam(url)
cities = get_cities(url)

c = []
print('Cities: ')
for i in cities:
    c.append(i.text)

print(c)

all_vals = []
for i in panchangam:
    all_vals.append(i.text)

today = datetime.date.today()
weekday_number = today.weekday()
weekday_name = datetime.date.today().strftime("%A")

title = all_vals[0]

col1 = ['Samvatsara']
col2 = ['Krodhi Nama Samvatsara']

skip = 0
for i in range(1, len(all_vals)):
    if all_vals[i] in ['Panchangam', 'Time to Avoid (Bad time to start any important work)', 'Good Time (to start any important work)']:
        continue
    elif skip % 2 != 0:
        col2.append(all_vals[i])
        skip += 1
    else:
        col1.append(all_vals[i])
        skip += 1

table = zip(col1, col2)


maps = {
    'Title': title,
}

for a, b in zip(col1, col2):
    if a == 'Month':
        maps['Masa'] = b
        maps['Ruthu'] = ruthu_map[b]
        maps['Vasara'] = day_map[weekday_name]
    else:
        if a in ['Tithi', 'Nakshatram', 'Yogam', 'Karanam']:
            b = b.replace('  ', '\n')
        maps[a] = b

details = f'City: {maps["City"]}, Sunrise: {maps["Sunrise"]}, Sunset: {maps["Sunset"]}'
maps['Details'] = details
maps.pop('City', None)
maps.pop('Sunrise', None)
maps.pop('Sunset', None)


for key, value in maps.items():
    print(f'{key}: {value}')
```
